````python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score
from sklearn.preprocessing import StandardScaler
import statsmodels.api as sm

plt.rcParams['figure.dpi'] = 120
plt.rcParams['figure.figsize'] = (10, 5)
sns.set_theme(style='whitegrid')

ALPHA = 0.05
````

## 1. Carregamento e pré-processamento

````python
df_raw = pd.read_csv('../data/dataset.csv')
print(f'Shape inicial: {df_raw.shape}')
print(f'Repos: {df_raw["repo"].nunique()} — {list(df_raw["repo"].unique())}')
df_raw.head(3)
````

````python
# Remove colunas de identificação/texto e colunas constantes (variância zero)
id_cols = ['class', 'file']
zero_var = [c for c in df_raw.columns if df_raw[c].nunique() <= 1 and c not in ['repo']]
print('Colunas constantes removidas:', zero_var)

df = df_raw.drop(columns=id_cols + zero_var, errors='ignore').copy()

# Métricas numéricas disponíveis para análise
METRICAS = [c for c in df.columns if c not in ['repo', 'Number_of_bugs']]
print('\nMétricas para análise:', METRICAS)
print(f'Shape final: {df.shape}')
````

````python
# Remove linhas com NaN nas métricas principais
df = df.dropna(subset=METRICAS).reset_index(drop=True)
print(f'Shape após remoção de NaN: {df.shape}')

# Triagem de alta correlação (Spearman > 0.95) — mantém a métrica mais interpretável
corr_matrix = df[METRICAS].corr(method='spearman').abs()
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
to_drop = [col for col in upper.columns if any(upper[col] > 0.95)]
print(f'Colunas altamente correlacionadas (|ρ| > 0.95) removidas: {to_drop}')

df = df.drop(columns=to_drop)
METRICAS = [c for c in df.columns if c not in ['repo', 'Number_of_bugs']]
print(f'Métricas finais ({len(METRICAS)}): {METRICAS}')
````

## 2. Estatística descritiva

````python
desc = df[METRICAS + ['Number_of_bugs']].describe(percentiles=[.25, .5, .75, .90, .95]).T
desc['amplitude'] = desc['max'] - desc['min']
desc = desc[['mean', '50%', '25%', '75%', '90%', '95%', 'max', 'amplitude', 'std']]
desc.columns = ['Média', 'Mediana', 'Q1', 'Q3', 'P90', 'P95', 'Máximo', 'Amplitude', 'DP']
desc.round(2)
````

````python
# Moda
modas = {c: df[c].mode().iloc[0] for c in METRICAS + ['Number_of_bugs']}
pd.Series(modas, name='Moda').to_frame()
````

````python
# Distribuição por repo
print('Classes por repositório:')
print(df.groupby('repo').size().sort_values(ascending=False).to_string())
````

## 3. Distribuição dos dados e testes de normalidade

````python
METRICAS_PRINCIPAIS = ['LOC', 'WMC', 'CBO', 'RFC', 'LCOM5', 'McCC_avg']
METRICAS_PRINCIPAIS = [m for m in METRICAS_PRINCIPAIS if m in df.columns]

n = len(METRICAS_PRINCIPAIS)
ncols = 3
nrows = (n + ncols - 1) // ncols

fig, axes = plt.subplots(nrows, ncols, figsize=(15, 4 * nrows))
axes = axes.flatten()

for i, col in enumerate(METRICAS_PRINCIPAIS):
    axes[i].hist(df[col], bins=40, edgecolor='white', color='steelblue', alpha=0.8, density=True)
    df[col].plot.kde(ax=axes[i], color='navy', linewidth=1.5)
    axes[i].set_title(col, fontsize=12)
    axes[i].set_xlabel('')
    axes[i].set_ylabel('Densidade')

for j in range(i + 1, len(axes)):
    axes[j].set_visible(False)

plt.suptitle('Histogramas com curva de densidade — Métricas principais', fontsize=14, y=1.01)
plt.tight_layout()
plt.savefig('normalidade.png', bbox_inches='tight')
plt.show()
````

````python
# Teste de Kolmogorov-Smirnov
ks_results = []
for col in METRICAS + ['Number_of_bugs']:
    data = df[col].dropna()
    stat, pval = stats.kstest(data, 'norm', args=(data.mean(), data.std()))
    ks_results.append({'Variável': col, 'Estatística D': round(stat, 4),
                        'p-valor': pval, 'Normal (α=0.05)': 'Não' if pval < ALPHA else 'Sim'})

pd.DataFrame(ks_results).set_index('Variável')
````

## 4. Identificação de outliers

````python
fig, axes = plt.subplots(1, len(METRICAS_PRINCIPAIS), figsize=(16, 5))

for i, col in enumerate(METRICAS_PRINCIPAIS):
    axes[i].boxplot(df[col].dropna(), vert=True, patch_artist=True,
                    boxprops=dict(facecolor='steelblue', alpha=0.6),
                    medianprops=dict(color='navy', linewidth=2),
                    flierprops=dict(marker='.', markersize=3, alpha=0.4))
    axes[i].set_title(col, fontsize=11)
    axes[i].set_xticks([])

plt.suptitle('Boxplots — Detecção de outliers (critério IQR)', fontsize=13)
plt.tight_layout()
plt.savefig('boxplot.png', bbox_inches='tight')
plt.show()

# Contagem de outliers por variável
print('\nOutliers por variável (método IQR):')
for col in METRICAS:
    q1, q3 = df[col].quantile([0.25, 0.75])
    iqr = q3 - q1
    n_out = ((df[col] < q1 - 1.5 * iqr) | (df[col] > q3 + 1.5 * iqr)).sum()
    pct = n_out / len(df) * 100
    print(f'  {col:15s}: {n_out:4d} outliers ({pct:.1f}%)')
````

## 5. Análise de correlação (Spearman)

````python
COLS_CORR = METRICAS + ['Number_of_bugs']
corr = df[COLS_CORR].corr(method='spearman')

mask = np.triu(np.ones_like(corr, dtype=bool))

fig, ax = plt.subplots(figsize=(12, 10))
sns.heatmap(corr, mask=mask, annot=True, fmt='.2f', cmap='RdYlGn',
            center=0, vmin=-1, vmax=1, ax=ax,
            annot_kws={'size': 8}, linewidths=0.5)
ax.set_title('Matriz de Correlação de Spearman', fontsize=14)
plt.tight_layout()
plt.savefig('correlacao.png', bbox_inches='tight')
plt.show()
````

````python
# Correlações com Number_of_bugs ordenadas
corr_bugs = corr['Number_of_bugs'].drop('Number_of_bugs').sort_values(key=abs, ascending=False)
print('Correlação de Spearman com Number_of_bugs:')
for var, val in corr_bugs.items():
    strength = 'forte' if abs(val) >= 0.70 else 'moderada' if abs(val) >= 0.40 else 'fraca'
    print(f'  {var:15s}: ρ = {val:+.3f}  ({strength})')
````

````python
# Gráficos de dispersão: métricas principais vs Number_of_bugs
SCATTER_COLS = [c for c in ['LOC', 'WMC', 'CBO', 'RFC'] if c in df.columns]

fig, axes = plt.subplots(1, len(SCATTER_COLS), figsize=(15, 4))
for i, col in enumerate(SCATTER_COLS):
    axes[i].scatter(df[col], df['Number_of_bugs'], alpha=0.3, s=15, color='steelblue')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Number_of_bugs' if i == 0 else '')
    r, p = stats.spearmanr(df[col], df['Number_of_bugs'])
    axes[i].set_title(f'ρ = {r:.3f} (p={p:.3f})')

plt.suptitle('Dispersão: métricas vs bugs', fontsize=13)
plt.tight_layout()
plt.savefig('dispersao.png', bbox_inches='tight')
plt.show()
````

## 6. Variável resposta e modelagem logística

````python
# Distribuição da variável resposta
fig, ax = plt.subplots(figsize=(6, 4))
counts = df['Number_of_bugs'].value_counts().sort_index()
ax.bar(counts.index.astype(str), counts.values, color='steelblue', edgecolor='white')
ax.set_xlabel('Number of bugs')
ax.set_ylabel('Frequência')
ax.set_title('Distribuição de Number_of_bugs')
plt.tight_layout()
plt.savefig('frequencia.png', bbox_inches='tight')
plt.show()

# Binarização
df['bug'] = (df['Number_of_bugs'] > 0).astype(int)
n_bug = df['bug'].sum()
n_total = len(df)
print(f'Sem bug: {n_total - n_bug} ({(n_total - n_bug)/n_total*100:.2f}%)')
print(f'Com bug: {n_bug} ({n_bug/n_total*100:.2f}%)')
````

````python
def fit_logistic(X_cols, df, label):
    """Ajusta regressão logística e retorna métricas de avaliação."""
    X_cols = [c for c in X_cols if c in df.columns]
    subset = df[X_cols + ['bug']].dropna()
    X = sm.add_constant(subset[X_cols])
    y = subset['bug']

    model = sm.Logit(y, X).fit(disp=False)

    # AUC
    probs = model.predict(X)
    auc = roc_auc_score(y, probs) if y.sum() > 0 else float('nan')

    # Pseudo R² de McFadden
    pseudo_r2 = model.prsquared

    return {
        'Modelo': label,
        'Preditores': ', '.join(X_cols),
        'AIC': round(model.aic, 2),
        'Pseudo R²': round(pseudo_r2, 4),
        'AUC': round(auc, 3),
        '_model': model,
    }

modelos_def = [
    (['LOC'],                             'Modelo 1'),
    (['WMC', 'CBO', 'RFC'],              'Modelo 2'),
    (['LOC', 'CBO', 'LCOM5'],            'Modelo 3'),
    (['LOC', 'WMC', 'CBO', 'RFC', 'LCOM5'], 'Modelo 4'),
]

resultados = [fit_logistic(cols, df, label) for cols, label in modelos_def]

tabela = pd.DataFrame([{k: v for k, v in r.items() if k != '_model'} for r in resultados])
tabela.set_index('Modelo')
````

````python
# Modelo com menor AIC
melhor = min(resultados, key=lambda r: r['AIC'])
print(f"Melhor modelo (menor AIC): {melhor['Modelo']} — AIC={melhor['AIC']}")
print()
print(melhor['_model'].summary2())
````

````python
# Odds Ratio do melhor modelo
m = melhor['_model']
or_df = pd.DataFrame({
    'Coeficiente (β)': m.params,
    'Odds Ratio': np.exp(m.params),
    'p-valor': m.pvalues,
    'Significativo': m.pvalues < ALPHA,
}).drop(index='const', errors='ignore').round(4)
or_df
````

## 7. Análise por repositório

````python
# Métricas médias por repo
metricas_repo = df.groupby('repo')[METRICAS].mean().round(2)
metricas_repo['total_classes'] = df.groupby('repo').size()
metricas_repo['classes_com_bug'] = df.groupby('repo')['bug'].sum()
metricas_repo['pct_bug'] = (metricas_repo['classes_com_bug'] / metricas_repo['total_classes'] * 100).round(2)
metricas_repo.sort_values('pct_bug', ascending=False)
````

````python
# Comparação visual: WMC e CBO médios por repo
cols_plot = [c for c in ['WMC', 'CBO', 'RFC', 'LOC'] if c in metricas_repo.columns]
fig, axes = plt.subplots(1, len(cols_plot), figsize=(16, 5))

repo_labels = [r.split('/')[-1] for r in metricas_repo.index]

for i, col in enumerate(cols_plot):
    axes[i].barh(repo_labels, metricas_repo[col], color='steelblue', alpha=0.8)
    axes[i].set_title(f'{col} médio por repo', fontsize=11)
    axes[i].invert_yaxis()

plt.suptitle('Comparação de métricas por repositório (média)', fontsize=13)
plt.tight_layout()
plt.savefig('repos_comparacao.png', bbox_inches='tight')
plt.show()
````
